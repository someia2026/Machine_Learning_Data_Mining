### Nettoyage Data

```python
valeurs_manquantes = df.isnull().sum
print(valeurs_manquantes)
```

#### Suppression de colonnes inutiles

```python
df = df.drop(columns=['CODE_PARENT','CODE_PARENT_DESC','VARIETE','EQUIPE', 'REMARQUES','RAISONDEPLANTATION','TRAITEMENTCHENILLES','COURRIER','IDENTIFIANTPLU','TYPEIMPLANTATIONPLU','INTITULEPROTECTIONPLU','ANNEEABATTAGE','ESSOUCHEMENT','DIAMETREARBRE', 'CAUSEABATTAGE'])
```

```python
colonnes_vide = df.isna().sum() / df.shape[0] * 100
colonnes_vide
```

#### Suppression des lignes des années de plantations sans valeurs

```python
df = df.dropna(subset=['ANNEEDEPLANTATION'])
df['ANNEEDEPLANTATION'].unique()
```

### Visualisation des data manquantes

```python
# Créer un tableau complet avec toutes les informations
missing_info = pd.DataFrame({
    'Colonne': df.columns,
    'Valeurs manquantes': df.isnull().sum().values,
    'Pourcentage (%)': (df.isnull().sum() / len(df) * 100).round(2).values,
    'Valeurs présentes': df.notnull().sum().values
})

print(missing_info)
```

```python
sns.heatmap(df.isnull(), cbar=False)
```

### Vérification après nettoyage

```python
colonnes_avec_nan = df.columns[df.isnull().any()].tolist()
print(colonnes_avec_nan)
```

```python
print(df.info())
```

```python
# Créer un tableau avec seulement les colonnes qui ont des NaN
missing = df.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)

if len(missing) > 0:
    missing_df = pd.DataFrame({
        'Colonne': missing.index,
        'Nombre de NaN': missing.values,
        'Pourcentage': (missing.values / len(df) * 100).round(2)
    })
    print(missing_df)
else:
    print("✓ Aucune colonne avec des NaN !")
```

#### Diviser la colonne GEOJson en deux distinctes (Latitude et Longitude)

```python
import json

# Nom de votre colonne (à ajuster)
colonne_geo = 'geo_point_2d'  # ou 'coordinates', 'geometry', etc.

# Extraire latitude et longitude
df['longitude'] = df['GeoJSON'].apply(lambda x: json.loads(x)['coordinates'][0])
df['latitude'] = df['GeoJSON'].apply(lambda x: json.loads(x)['coordinates'][1])

print(df[['longitude', 'latitude']].head())

```
